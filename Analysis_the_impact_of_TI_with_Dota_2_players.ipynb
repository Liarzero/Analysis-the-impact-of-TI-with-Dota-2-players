**See only game DOTA 2**

```python
Dota_2 = data[data['Game_Name'] == 'Dota 2'].sort_values(by="Date")
```

```python
plt.figure(figsize=(70,10))
plt.subplot(1, 2, 2)
sns.lineplot(data=Dota_2,  x="Month_Year", y="Avg_players")
plt.xticks(rotation=90)
plt.title('Average Players of Dota 2 All The Time')
```

This is the graphic of Average Dota 2 Players All The Time

```python
plt.figure(figsize=(70,10))
plt.subplot(1, 2, 1)
sns.lineplot(data=Dota_2,  x="Month_Year", y="Peak_Players")
plt.xticks(rotation=90)
plt.title('Peak Player of DOTA 2 All The Time')
```

This is the graphic of peak players Dota 2 All The Time

```python
plt.figure(figsize=(70,10))
plt.subplot(1, 2, 2)
sns.lineplot(data=Dota_2,  x="Month_Year", y="Avg_players")
plt.xticks(rotation=90)
plt.title('Average Players of Dota 2 After Every TI Since TI 2')
plt.axvline(x="August 2013")
plt.axvline(x="July 2014")
plt.axvline(x="August 2015")
plt.axvline(x="August 2016")
plt.axvline(x="August 2017")
plt.axvline(x="August 2018")
plt.axvline(x="August 2019")
```

This is the graphic of average players Dota 2 after every TI since TI 2

```python
plt.figure(figsize=(70,10))
plt.subplot(1, 2, 1)
sns.lineplot(data=Dota_2,  x="Month_Year", y="Peak_Players")
plt.xticks(rotation=90)
plt.title('Peak Players of Dota 2 After Every TI Since TI 2')
plt.axvline(x="August 2013")
plt.axvline(x="July 2014")
plt.axvline(x="August 2015")
plt.axvline(x="August 2016")
plt.axvline(x="August 2017")
plt.axvline(x="August 2018")
plt.axvline(x="August 2019")
```

This is the graphic of peak players Dota 2 after every TI since TI 2
